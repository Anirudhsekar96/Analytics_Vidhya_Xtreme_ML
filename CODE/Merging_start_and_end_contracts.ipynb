```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("final_indicacores_nona.csv")
df2 = pd.read_csv("Contracts_End.csv")
df3 = pd.read_csv("Contracts_New.csv")
```

```python
df_end = df2.merge(df, left_on='CENSUS_SECTION', right_on='aa_combined', how='outer')
```

```python
df_new = df3.merge(df, left_on='CENSUS SECTION', right_on='aa_combined', how='outer')
```

```python
df_end.to_csv("end_combined.csv",index=False)
df_new.to_csv("new_combined.csv",index=False)
```

```python
df_end2 = df2.merge(df, left_on='CENSUS_SECTION', right_on='aa_combined', how='inner')
df_new2 = df3.merge(df, left_on='CENSUS SECTION', right_on='aa_combined', how='inner')
df_end2.to_csv("end_combined_final.csv",index=False)
df_new2.to_csv("new_combined_final.csv",index=False)
```
