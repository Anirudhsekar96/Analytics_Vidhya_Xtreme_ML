```python
# import modules
import pandas as pd

# load data files
contacts_pre_2017=pd.read_csv('Contacts_Pre_2017.csv')
resolution_pre_2017=pd.read_csv('Resolution_Pre_2017.csv', encoding='iso8859-1')

contacts2017=pd.read_csv('Contacts2017.csv')
resolution2017=pd.read_csv('Resolution2017.csv')

# create average contact on the basis of contact type
Contacts_Avg=pd.DataFrame({'Contacts' : contacts_pre_2017.groupby(['CONTACT.TYPE'] )['Contacts'].mean()}).reset_index()
contacts2017.drop('Contacts', axis=1, inplace=True)
Contacts=pd.merge(contacts2017, Contacts_Avg, how='left', on=['CONTACT.TYPE'])

# create average resolution on the basis of category and subject
Resolution_Avg=pd.DataFrame({'Resolution' : resolution_pre_2017.groupby(['Category','Subject'] )['Resolution'].mean()}).reset_index()
resolution2017.drop('Resolution', axis=1, inplace=True)
Resolution=pd.merge(resolution2017, Resolution_Avg, how='left', on=['Category','Subject'])
Resolution['Resolution'].fillna(1, inplace=True)

# create submission file
Resolution[['ID', 'Resolution']].to_csv('Resolution2_benc.csv', index=False)
Contacts[['ID', 'Contacts']].to_csv('Contacts_benc.csv', index=False)
```
